### Treating the outliers

```python
# we will check the outliers and we will find out the parcengtage of data lose 
# we will remove the outliers depending on the data loss parcentage
# we can't loos data more than 5% at any point. 
```

```python
# we will check the outlier using box plot class wise
plt.figure(figsize = (8,4) , dpi = 100) # this line of code for more clearer image
sns.boxplot(x='class',y='amount',data=df)
plt.show()
```

```python
# the image shows distribution of data class wise, 
# here we can easyly understand that there are more outliers in geniune transactions
# fraud transactions has less number of outliers 
# The amount of fraud transaction are less than 5000 euros.
```

```python
# our aim is to detect the fraud transaction 
# in our data fraud transactions has only 0.1727% which is very very low in numbers.
# keeping it in mind we should not loose any data from the fraud transaction side.
# so in outlier removal we will set our upper bound to the maximum value of fraud transactions.
# but we will also look at the data that how much % of data we loose if we remove the outlier using contemporary method.
```

```python
# we will write a function to find out the outliers using contemporary method. 
def indicies_of_outliers(x):
    q1,q3=np.percentile(x,[25,75])
    iqr=q3-q1
    lower_bound=q1-(iqr*1.5)
    upper_bound=q3+(iqr*1.5)
    
    return np.where((x > upper_bound)|(x < lower_bound))
```

```python
indicies_of_outliers(df['amount'])
# this are the index value of the outliers
```

```python
# counting how many outliers are there
len(indicies_of_outliers(df['amount'])[0])
```

```python
# total how many data points we have
len(df)
```

```python
# lets check what parcentage of data we will loose if we just directly remove the outlier
len(indicies_of_outliers(df['amount'])[0])/len(df)
```

```python
# so if we drop outliers directly then we loose 11.18% of the data
# we could end up loosing outlier fraud transaction. which we can't effort to do.
# a good number of outlier removed would be 5-6 % 
```

```python
# in case of an unbalanced data the proportion of 0 & 1 should be similar
# before and after the removal of outlier.
# lets check the propotion before and after removing the outliers
```

```python
# propotion of 0 & 1 before removal of outlier
df['class'].value_counts(normalize=True)
```

```python
# propotion of 0 & 1 after removal of outliers
df[df.index.isin(list(indicies_of_outliers(df['amount'])[0]))]['class'].value_counts(normalize = True)
```

```python
# there is a difference betweent the before and after removal proportion of 0 & 1 values
# but this difference is not significent, but in this way we will loose the fraud transaction ouliers as well
# and our data is already short form the fraud side.
# so we can't loos data from the fraud side
```
